```python
# Calculate average order value and total orders per restaurant
restaurant_stats = (
    final_df
    .groupby("restaurant_name_x")["total_amount"]
    .agg(avg_order_value="mean", total_orders="count")
)

# Filter restaurants with less than 20 orders
filtered_restaurants = restaurant_stats[restaurant_stats["total_orders"] < 20]

# Get restaurant with highest average order value
filtered_restaurants.sort_values(
    by="avg_order_value",
    ascending=False
).head(1)
```
